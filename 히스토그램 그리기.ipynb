```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
# alpha는 투명도, density는 y축을 빈도수(frequency)가 아니라 0~1 범위로 지
plt.hist(iris['sepal_width'], bins=10, density=True, 
         facecolor='blue', alpha=0.5)
plt.xlabel('X bins')
plt.ylabel('Density')
plt.title('Histogram of Sepal Width')
plt.axis([1.5, 4.5, 0, 1.1])
plt.show()
```

```python
# seaborn을 통한 히스토그램 그리기
sns.distplot(iris['sepal_width'], hist=True, kde=True, bins=10, color='blue', 
             hist_kws={'edgecolor': 'gray'}, kde_kws={'linewidth': 2})
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width(cm)')
plt.ylabel('Density')
plt.show()
```

```python
# pandas를 통한 히스토그램 그리기
iris['sepal_width'].hist(bins=10, grid=False)
plt.show()
```
