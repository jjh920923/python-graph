```python
#annotate tutorial
#https://www.geeksforgeeks.org/zip-in-python/
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10,1)
y = np.random.normal(loc=2, scale=.8, size=10)
plt.plot(x,y)
plt.text(2,4,'this text starts at point (2,4)')
plt.text(8,3, 'this text ends at point (8,3)', horizontalalignment='right')
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,5,.5))
plt.show()
```

```python
x = np.arange(0,10,1)
y = np.random.normal(loc=3, scale=.4, size=10)
plt.plot(x,y,'bo-')
for x,y in zip(x,y):
    label = "{:.2f}".format(y)
    plt.annotate(label,
                 (x,y),
                 textcoords='offset points',
                 xytext=(0,10),
                 ha='center')
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,7,.5))
plt.show()
```

```python
x = np.arange(0,10,1)
y = np.random.normal(loc=3, scale=.4, size=10)
plt.bar(x,y)
for x,y in zip(x,y):
    label = "{:.2f}".format(y)
    plt.annotate(label,
                 (x,y),
                 textcoords='offset points',
                 xytext=(0,10),
                 ha='center')
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,5,.5))
plt.show()
```

```python
x = np.random.normal(loc=4, scale=2, size=10)
y = np.random.normal(loc=2, scale=.8, size=10)
plt.scatter(x,y)
for x,y in zip(x,y):
    label = "{:.2f}".format(y)
    plt.annotate(label,
                 (x,y),
                 textcoords='offset points',
                 xytext=(0,10),
                 ha='center')
plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,5,.5))
plt.show()
```
