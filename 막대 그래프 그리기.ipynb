```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips_sum_by_day = tips.groupby('day').tip.sum()
tips_sum_by_day
```

```python
tips_sum_by_day.index
label = tips_sum_by_day.index
index = np.arange(len(label))
```

```python
# 그냥 barplot을 그리면 x의 범위가 원하는 대로 안나오기 때문에
# xticks를 이용하여 지정해주어야 한다.
plt.bar(index, tips_sum_by_day, color='red', alpha=0.5, width=0.5)
plt.title('Sum of Tips by Day', fontsize=20)
plt.xlabel('Day', fontsize=18)
plt.ylabel('Sum of Tips', fontsize=18)
plt.xticks(index, label, rotation=60, fontsize=15)
plt.show()
```

```python
# 가로로 그래프를 보고싶다면 barh를 사용한다.
plt.barh(index, tips_sum_by_day, color='red', alpha=0.5, height=0.5)
plt.title('Sum of Tips by Day', fontsize=20)
plt.xlabel('Day', fontsize=18)
plt.ylabel('Sum of Tips', fontsize=18)
plt.yticks(index, label, rotation=60, fontsize=15)
plt.show()
```

```python
# 겹친 막대 그래프를 그리기 위한 데이터 변환
tips_sum_by_day_sex = pd.DataFrame(tips.groupby(['day', 'sex']).tip.sum())
tips_sum_by_day_sex = tips_sum_by_day_sex.reset_index()
tips_sum_by_day_sex
```

```python
tips_sum_by_day_sex = tips_sum_by_day_sex.sort_values(by=['day', 'sex'], ascending=False)
tips_sum_by_day_sex
```

```python
tips_sum_by_day_sex['tip_cumsum'] = tips_sum_by_day_sex.groupby(['day'])['tip'].cumsum(axis=0)
tips_sum_by_day_sex
```

```python
tips_sum_by_day_sex = tips_sum_by_day_sex.sort_values(by=['day', 'sex'], ascending=True)
tips_sum_by_day_sex
```

```python
# dodge 명령어를 통해 그래프를 쌓을 것인지 결정할 수 있다.
sns.barplot(x='day', y='tip_cumsum', hue='sex', data=tips_sum_by_day_sex, dodge=False)
plt.title('Stacked Bar Chart by Seaborn', fontsize=20)
plt.show()
```

```python
sns.barplot(x='day', y='tip', hue='sex', data=tips_sum_by_day_sex)
plt.title('Dodged Bar Chart by Seaborn', fontsize=20)
plt.show()
```
